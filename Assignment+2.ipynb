1. Create the below pattern using nested for loop in Python. 

```python
for i in range(0,6):
    print("*"*i)
for i in range(4,0,-1):
    print("*"*i)
```

Alternative using for and while loop

```python
for i in range(0,6):
    print("*"*i)
i=4
while i <= 4:
    print("*"*i)
    i-=1
```

2. Write a Python program to reverse a word after accepting the input from the user.
Input word: ineuron 
Output: norueni
